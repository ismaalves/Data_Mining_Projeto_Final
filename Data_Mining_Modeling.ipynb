```python
import pandas as pd
# Carregar os dados
df = pd.read_csv('/content/drive/MyDrive/data/Preenchido/Dados_Preenchidos.csv')
df = df.drop(columns=['Data'])
```

```python
df.head()
```

```python
# Tabela Fato - Coleta de Impostos
fato_impostos = df[['Cod.IBGE', 'Mês', 'Ano', 'Conta', 'Valor', 'População']]
fato_impostos.head()
```

```python
# Dimensão Tempo
dim_tempo = df[['Mês', 'Ano']].drop_duplicates().reset_index(drop=True)
dim_tempo['Chave_Tempo'] = dim_tempo.index + 1
dim_tempo.head()
```

```python
# Dimensão Geografia
dim_geografia = df[['Cod.IBGE', 'Instituição', 'UF']].drop_duplicates().reset_index(drop=True)
dim_geografia.head()
```

```python
# Dimensão Conta (Tipo de Imposto)
dim_conta = df[['Conta']].drop_duplicates().reset_index(drop=True)
dim_conta['Chave_Conta'] = dim_conta.index + 1
dim_conta.head()
```

```python
# Agora você pode adicionar as chaves estrangeiras na tabela fato
fato_impostos = fato_impostos.merge(dim_tempo, on=['Mês', 'Ano'], how='left')
fato_impostos = fato_impostos.merge(dim_geografia, on='Cod.IBGE', how='left')
fato_impostos = fato_impostos.merge(dim_conta, on='Conta', how='left')
```

```python
# Selecionando apenas as chaves e a medida na tabela fato
fato_impostos = fato_impostos[['Cod.IBGE', 'Chave_Tempo', 'Chave_Conta', 'Valor', 'População']]
```

```python
fato_impostos.to_csv('/content/drive/MyDrive/data/Data_Tax.csv', index=False)
dim_tempo.to_csv('/content/drive/MyDrive/data/Data_Time.csv', index=False)
dim_geografia.to_csv('/content/drive/MyDrive/data/Data_Geo.csv', index=False)
dim_conta.to_csv('/content/drive/MyDrive/data/Data_TaxType.csv', index=False)
```
